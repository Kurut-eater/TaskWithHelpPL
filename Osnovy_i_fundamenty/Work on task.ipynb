```python
m4=[1.25,1.8,2.5,3.5,5,7.5] #Масса ударной части молота
Bt=500 #число ударов молота, необходимое для погружения свай, принимаем равным

n=5.5 #параметр, при трубчатых дизель-молотах

#3) Значение необходимой энергии удара молота Ел, кДж, обеспе-
#чивающей погружение свай до проектной отметки без дополнитель-
#ных мероприятий, следует определять по формуле
for show in (
    (i+1, f"{Eh}>={round((Fd/Bt)*(n+(mc/m4[i])),12)}",
    answer[Eh>=(Fd/Bt)*(n+(mc/m4[i]))]) 
    for i in range(len(m4))):
    print(show)
```

```python
A=40 #площадь поперечного сечения сваи, m**2
nyu=1500 #коэффициент материала сваи, принимаемый для железобетонных свай 

#4) Значение контрольного остаточного Sa, м, отказа при забивке
#и забивке свай в зависимости от энергии удара выбранного молота
#и несущей способности сваи F4, указанной в проекте, должно удовле-
#творять условию
for show in [(i+1,f"{Sa}<={((nyu*A*Ed[i])    / (Fd*(Fd+nyu*A)))
                            *((m1[i]+0.2*m2) / (m1[i]+m2))}",
        answer[
Sa<=((nyu*A*Ed[i])/
     (Fd*(Fd+nyu*A)))
              *
    ((m1[i]+0.2*m2)/
     (m1[i]+m2))
        ]
)for i in range(6)]:
    print(show)
```

```python
    #Выбор типа вибропогружателя для погружения свайных элементов
#Данные о грунтах основания, размерах, массе и несущей способности сваи
#песок пылеватый рыхлый 
Fd=3100#кН Несущая способность свай
mc=54#тс Масса свай с наголовником
l=20#м длина свай
d=3#м Диаметр свай
    #Тип вибропрогужателя
    #ВРП70/200
#Техническая характеристика вибропогружателей (Россия)
K_m=[2.3,7]#кН*м  Статический момент массы дебалансов
n__д=[0,500]# Частота вращения дебалансов
n__w=[240,500] # Рабочая частота 
Pb=2000#кН  Вынуждающая сила максимальная
W_h=200 #Мощность основного электродвигателя,
m1=14#Масса вибропогружателя без наголовника
#Габариты
size=[3250,1700,1346]# высота, длина, ширина
dives=2 # сечения погружаемых свай
dives_until=40 # глубина погружения до
#+
ro_k=1.4 #коэффициент надежности по грунту, принимаемый равным
N=Fd/ro_k #расчетная нагрузка на свайный элемент
g1,gc=[i*9.81 for i in (m1,mc)] # вес вибропогружателя,вес сваи с наголовником

for show in (f"N = {N}",f"g1 = {g1}",f"gc = {gc}"):
 print(show)
Ks=3.5# коэффициент снижения бокового сопротивления грунта во
    # время вибропогружения, принимаемый по табл.8.4 для песчаных
    # грунтов и по табл. 8.5 для глинистых грунтов.
#1) Значение необходимой минимальной вынуждающей силы виб-
#ропогружателя Fn, кН, определяют по формуле
F_0=(ro_k*N-2.8*(g1+gc))/Ks
print("F_0 =",F_0)
```

```python
A_h=(1.4,1.6)#необходимая амплитуда колебаний при отсутствии
             # сопротивлений грунта, см, принимаемая по табл. 8.7.
#Условия принятия тип молота
for show in[(f"{K_m[i]}>={(m1+mc)*A_h[i]}",answer[K_m[i]>=(m1+mc)*A_h[i]]) for i in (0,1)]:
    print(show)
#!разница может быть большым!
```

```python
V=2 # В конце вибропогружения висячего свайного элемента при
    # скорости вибропогружения в последнем залоге не менее 2 см/мин
fr=1 # коэффициент влияния инерционных и вязких сопротивлений на несущую способность сваи,
     # принимаемый по табл. 8.8; моем случие - Пески и супеси твердые
ro_g=1.4 #коэффициент надежности по грунту
W=0.9*W_h-W_h/4 # мощность, расходуемая на движение вибросистемы, кВт W=nyu*Wh-W0
m_dc=m1+0.25#m1+0.25 # Масса вибросистемы
A_0=[100*i/m_dc for i in K_m]# расчетная амплитуда колебаний вибросистемы без сопротив-
                             # линий, см, вычисляемая по формуле
Ar=[i/2 for i in A_0]# фактическая амплитуда колебаний, принимаемая равной половине
                     # полного размаха колебаний свайного элемента на последней минуте погружения, см;
# боковое сопротивление грунта при вибропогружении
Fs=[(1.5*10**3*W)/(Ar[i]*(n__д[i]+((V+2)/2*A_0[i]))) for i in (0,1)]# кН
print("W =",W,
      "\nA_0 =",A_0,
      "\nAr =",Ar,
      "\nFs =",Fs)
# 2) Условия
for show in [(f"{N} <= {((6000*W-2*n__д[i]*Fs[i]*
             (2*Ar[i]-V/# при делени на 0 выходит ощибка
              (0.00001 if n__д[i]==0 else n__д[i])#если n__d ровна 0 то вместо n__d вставится 0.00001
         ))/V+Fs[i]**Ks-1)+(g1+gc)*(fr/ro_g)}",
        answer[N<=((6000*W-2*n__д[i]*Fs[i]*(2*Ar[i]-V/(0.00001 if n__д[i]==0 else n__д[i])))
                   /V+Fs[i]**Ks-1)+(g1+gc)*(fr/ro_g)])
       for i in (0,1)]:
    print(show)
```
